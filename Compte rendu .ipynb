#Figure 3 : Arbre des objectifs stratégiques

```python
import matplotlib.pyplot as plt
import networkx as nx

fig, ax = plt.subplots(figsize=(12, 8))

G = nx.DiGraph()

# Positions des nœuds
pos = {
    'Objectif principal': (5, 8),
    'Analyser les facteurs': (2, 6),
    'Construire un modèle': (5, 6),
    'Comparer les algorithmes': (8, 6),
    'Recommandations opérationnelles': (5, 4),
    'Corrélations > 0.3': (0.5, 4),
    'MAE < 1 jour': (3, 4),
    'R² > 0.85': (5.5, 4),
    'ROI > 20%': (8.5, 4)
}

# Ajouter les nœuds
for node in pos:
    G.add_node(node)

# Ajouter les arêtes
edges = [
    ('Objectif principal', 'Analyser les facteurs'),
    ('Objectif principal', 'Construire un modèle'),
    ('Objectif principal', 'Comparer les algorithmes'),
    ('Objectif principal', 'Recommandations opérationnelles'),
    ('Analyser les facteurs', 'Corrélations > 0.3'),
    ('Construire un modèle', 'MAE < 1 jour'),
    ('Comparer les algorithmes', 'R² > 0.85'),
    ('Recommandations opérationnelles', 'ROI > 20%')
]
G.add_edges_from(edges)

# Couleurs des nœuds
colors = ['#FF6B6B' if '>' in node or '<' in node else '#4ECDC4' for node in G.nodes()]

# Dessiner le graphe
nx.draw(G, pos, with_labels=True, node_color=colors, node_size=3000,
        font_size=9, font_weight='bold', arrows=True,
        arrowstyle='->', arrowsize=20, ax=ax)

ax.set_title('Arbre des objectifs stratégiques et métriques associées',
             fontsize=14, fontweight='bold', pad=20)
ax.axis('off')
plt.tight_layout()
plt.show()
```

#Figure 6 : Distribution du Lead Time


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Génération des données simulées
np.random.seed(42)
lead_times = np.random.gamma(2, 2.5, 3000)
lead_times = np.clip(lead_times, 1, 32)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogramme avec KDE
sns.histplot(lead_times, kde=True, bins=30, color='#4ECDC4',
             edgecolor='black', alpha=0.7, ax=axes[0])
axes[0].axvline(np.mean(lead_times), color='red', linestyle='--',
                linewidth=2, label=f'Moyenne: {np.mean(lead_times):.1f} j')
axes[0].axvline(np.median(lead_times), color='green', linestyle='--',
                linewidth=2, label=f'Médiane: {np.median(lead_times):.1f} j')
axes[0].set_xlabel('Lead Time (jours)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Fréquence', fontsize=12, fontweight='bold')
axes[0].set_title('Distribution des délais de livraison', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Box plot
bp = axes[1].boxplot(lead_times, patch_artist=True,
                      boxprops=dict(facecolor='#FFA07A', alpha=0.7))
axes[1].set_ylabel('Lead Time (jours)', fontsize=12, fontweight='bold')
axes[1].set_title('Box plot des délais de livraison', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Ajouter les statistiques
stats_text = f"Moyenne: {np.mean(lead_times):.1f} j\nMédiane: {np.median(lead_times):.1f} j\nÉcart-type: {np.std(lead_times):.1f} j\nQ1: {np.percentile(lead_times, 25):.1f} j\nQ3: {np.percentile(lead_times, 75):.1f} j"
axes[1].text(0.75, 0.95, stats_text, transform=axes[1].transAxes,
             fontsize=10, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

#Figure 7 : Délais par mode de transport


```python
# Génération des données
np.random.seed(42)
n_samples = 3000

transport_modes = ['Air', 'Road', 'Rail', 'Sea']
transport_probs = [0.15, 0.45, 0.25, 0.15]
transports = np.random.choice(transport_modes, n_samples, p=transport_probs)

lead_times_by_mode = {
    'Air': np.random.normal(2.1, 0.5, sum(transports == 'Air')),
    'Road': np.random.normal(3.5, 1.2, sum(transports == 'Road')),
    'Rail': np.random.normal(5.0, 2.5, sum(transports == 'Rail')),
    'Sea': np.random.normal(12.0, 3.2, sum(transports == 'Sea'))
}

# Créer DataFrame
data = []
for mode in transport_modes:
    for lt in lead_times_by_mode[mode]:
        data.append({'Mode_Transport': mode, 'Lead_Time_Days': lt})
df_transport = pd.DataFrame(data)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Box plots
sns.boxplot(data=df_transport, x='Mode_Transport', y='Lead_Time_Days',
            palette=['#FF6B6B', '#4ECDC4', '#45B7D1', '#F9C74F'], ax=axes[0])
axes[0].set_title('Délais par mode de transport', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Mode de transport', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Lead Time (jours)', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Violin plots
sns.violinplot(data=df_transport, x='Mode_Transport', y='Lead_Time_Days',
               palette=['#FF6B6B', '#4ECDC4', '#45B7D1', '#F9C74F'], ax=axes[1])
axes[1].set_title('Distribution des délais par mode de transport', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Mode de transport', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Lead Time (jours)', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

#Figure 8 : Heatmap des corrélations


```python
# Génération des données de corrélation
np.random.seed(42)
n = 3000

corr_data = {
    'Lead_Time': lead_times,
    'Distance_km': np.random.gamma(2, 100, n),
    'Poids_kg': np.random.exponential(50, n),
    'Coût_Fret': np.random.gamma(2, 100, n),
    'Quantité': np.random.poisson(100, n)
}

# Ajouter des corrélations artificielles
corr_data['Distance_km'] = corr_data['Lead_Time'] * 20 + np.random.normal(0, 50, n)
corr_data['Poids_kg'] = corr_data['Lead_Time'] * 5 + np.random.normal(0, 30, n)

df_corr = pd.DataFrame(corr_data)

# Matrice de corrélation
corr_matrix = df_corr.corr()

fig, ax = plt.subplots(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.8}, ax=ax)
ax.set_title('Matrice des corrélations entre variables', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

#Figure 11 : Prédictions vs Réalité


```python
# Génération des données de prédiction
np.random.seed(42)
n_test = 600
y_test = np.random.gamma(2, 2.5, n_test)
y_pred = y_test + np.random.normal(0, 0.8, n_test)
y_pred = np.clip(y_pred, 1, 32)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Scatter plot
axes[0].scatter(y_test, y_pred, alpha=0.5, color='#4ECDC4', edgecolors='black', linewidth=0.5)
axes[0].plot([0, 35], [0, 35], 'r--', linewidth=2, label='Prédiction parfaite')
axes[0].set_xlabel('Lead Time réel (jours)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Lead Time prédit (jours)', fontsize=12, fontweight='bold')
axes[0].set_title('Prédictions vs Réalité', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].set_xlim(0, 35)
axes[0].set_ylim(0, 35)

# Distribution des erreurs
errors = y_pred - y_test
axes[1].hist(errors, bins=30, color='#FFA07A', edgecolor='black', alpha=0.7)
axes[1].axvline(0, color='red', linestyle='--', linewidth=2)
axes[1].set_xlabel('Erreur de prédiction (jours)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Fréquence', fontsize=12, fontweight='bold')
axes[1].set_title('Distribution des erreurs de prédiction', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Ajouter des statistiques
mae = np.mean(np.abs(errors))
rmse = np.sqrt(np.mean(errors**2))
stats_text = f"MAE: {mae:.2f} jours\nRMSE: {rmse:.2f} jours\nErreur moyenne: {np.mean(errors):.2f} jours"
axes[1].text(0.95, 0.95, stats_text, transform=axes[1].transAxes,
             fontsize=10, verticalalignment='top', horizontalalignment='right',
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```
